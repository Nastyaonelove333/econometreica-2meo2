```python
import numpy as np # linear algebra 
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
# stats models: regression fitting
import statsmodels.formula.api as smf
# data visualization
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/artamonoff/Econometrica/master/python-notebooks/data-csv/sleep75.csv')
df
```

```python
df.shape
```

# 2.1.1

```python
sns.scatterplot(data=df, y='sleep', x='totwrk')
```

```python
sns.regplot(data=df, y='sleep', x='totwrk', ci=None, line_kws={"color": "hotpink"})
```

```python
# подгонка прямой
fitted_line = smf.ols(formula='sleep~totwrk', data=df).fit()
# коэффициенты с округлением
fitted_line.params.round(2)
```

## 2.1.2

```python
sns.regplot(data=df, y='sleep', x='age', ci=None, line_kws={"color": "lime"})
```

```python
# подгонка прямой
fitted_line = smf.ols(formula='sleep~age', data=df).fit()
# коэффициенты с округлением
fitted_line.params.round(2)
```

## 2.1.3

```python
sns.regplot(data=df, y='sleep', x='totwrk', ci=None, order=2, line_kws={"color": "tomato"})
```

```python
# подгонка параболы
fitted_pol = smf.ols(formula='sleep~totwrk+I(totwrk**2)', data=df).fit()
# коэффициенты с округлением
fitted_pol.params.round(2)
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/artamonoff/Econometrica/master/python-notebooks/data-csv/Labour.csv')
df
```

```python
sns.regplot(data=df, x='capital', y='output', ci=None, line_kws={"color": "yellow"})
```

```python
# подгонка параболы
fitted_pol = smf.ols(formula='output~capital', data=df).fit()
# коэффициенты с округлением
fitted_pol.params.round(2)
```

```python
sns.regplot(data=df, y=np.log(df['output']), x=np.log(df['capital']), ci=None, line_kws={"color": "y"})
```

```python
# подгонка прямой
fitted_line = smf.ols(formula='np.log(output)~np.log(capital)', data=df).fit()
# коэффициенты с округлением
fitted_line.params.round(2)
```
